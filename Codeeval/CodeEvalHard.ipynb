```python
def point_inside_polygon(x,y,poly):

    n = len(poly)
    inside =False

    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if y > min(p1y,p2y):
            if y <= max(p1y,p2y):
                if x <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= xinters:
                        inside = True
        p1x,p1y = p2x,p2y

    return inside

import re
with open(input(), "r") as cef:
    for lines in cef:
        alls, target = lines.split("|")
        x,y = map(int, target.strip().split())
        
        pcords = [x for x in alls.split(",")]
        poly = []
        for recs in pcords:
            rec = recs.strip().split(" ")
            poly.append((int(rec[0]),int(rec[1])))
        res = point_inside_polygon(x, y, poly)
        print("Prisoner" if res else "Citizen")
            
            
            
            
            
            
       
        
        
```

```python
import sys
from operator import itemgetter
import math

def getnearest(points):
    
    max_distance = 10000 * 10000 + 1
    ls = sorted(points, key=itemgetter(0, 1))    
    distance = max_distance
    cnt = len(ls)
    ds = 0
    for i in range(cnt):
        for j in range(i+1, cnt):
            x1 = ls[i][0]
            x2 = ls[j][0]
            xd = x2 - x1
            if xd * xd > max_distance:
                continue
            y1 = ls[i][1]
            y2 = ls[j][1]
            yd = y2 - y1 
            if yd * yd > max_distance:
                continue
            ds = xd * xd  + yd * yd            
            if ds < distance:
                distance = ds
    distance = math.sqrt(distance)    
    print("INFINITY" if distance > 10000 else "%.4f" % distance )            
               
   
        
cef = open(input(), "r")
toContinue = True
while toContinue:
    line = cef.readline().strip()    
    if line == '0':
        toContinue = False
        break
    points = []    
    for i in range(int(line)):
        points.append(list(map(int, cef.readline().strip().split())))        
    getnearest(points)
        
    
        
```
